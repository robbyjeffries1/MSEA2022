<font color = 'blue' size = '3'>**Plotting a White Noise Process**<br /><br />
  &emsp;Based on:  https://python-programming.quantecon.org/python_by_example.html<br /><br />
Objective:  Simulate and plot a white noise process $\epsilon_0, \epsilon_1, ..., \epsilon_T$, where each $\epsilon_t$ is an indepdendent and identically distributed (IID) draw from a Normal distribution

```python
# Load resources

import numpy as np                # NumPy works with arrays, contains common math functions, generates random numbers, etc.
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (14, 6)
```

<br /><font color = 'blue' size = '3'>**Random Seed**<br /><br />
A seed value is the base value used by a pseudo-random generator to produce random numbers<br />
The function 'random.seed()' initializes the pseudo-random number generator in Python; by default the random number generator uses the current system time as the seed<br />
*Setting* the seed value to a specific value re-uses the seed value, allowing draws to be reproduced by a pseudo-random number generator<br /><br />
    
In the examples below, setting the seed by specifying 'np.random.seed(0)' generates the same set of random numbers for each method - with the same random draws, the plots are the same<br /><br />
    
Note: Because they depend on a seedm 'random' numbers are only 'pseudo random' numbers 

<br /><font color = 'blue' size = '3'>**Version 1:  Direct Approach**<br />
    This approach uses the NumPy function 'random' to make draws from a normal distribution ('randn')

```python
np.random.seed(0)

_values = np.random.randn(100)   # Use the NumPy subpackage 'random' to generate
                                  # 100 independent standard normals and stores them in '_values'

plt.plot(_values)

plt.title('\nWhite Noise Process - Direct Approach\n', fontsize = 18)
plt.xlabel('time ($t$)', fontsize = 16, loc = 'right')
plt.ylabel('$\epsilon_t$', fontsize = 18, loc = 'top')

plt.show()
```

<br /><font color = 'blue' size = '3'>**Version 2:  Random Draws using a 'for' loop**<br />
The for loop is the most common technique for iteration in Python

```python
np.random.seed(0)

ts_length = 100  # Set length of the time series

_values = []    # Create an empty list that will store the draws
                 # A list is a native Python data structure used to group a collection of objects

for i in range(ts_length):   # 'range(ts_length)' defines a sequence over which the for loop will iterate
                             # The following two indented lines (a code block) are executed ts_length times before moving on
    e = np.random.randn()      # Make a draw
    _values.append(e)         # Add the draw to the list of previous draws

plt.plot(_values)

plt.title('\nWhite Noise Process Generated with a for Loop\n', fontsize = 18)
plt.xlabel('time ($t$)', fontsize = 16, loc = 'right')
plt.ylabel('$\epsilon_t$', fontsize = 18, loc = 'top')

plt.show()
```

<br /><font color = 'blue' size = '3'>**Version 3:  Random Draws using a 'while' Loop**<br />

```python
np.random.seed(0)

ts_length = 100

_values = []

i = 0   # Initialize a counter variable

while i < ts_length:         # As long as i is less than ts_length, perform another iteration
    e = np.random.randn()
    _values.append(e)
    i = i + 1                # Update the counter after each iteration
                             # 'i = i + 1' and 'i += 1' are equivalent
    
plt.plot(_values)

plt.title('\nWhite Noise Process Generated with a while Loop\n', fontsize = 18)
plt.xlabel('time ($t$)', fontsize = 16, loc = 'right')
plt.ylabel('$\epsilon_t$', fontsize = 18, loc = 'top')

plt.show()
```
