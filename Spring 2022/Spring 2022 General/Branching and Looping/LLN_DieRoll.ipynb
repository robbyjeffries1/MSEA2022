<font color = 'blue' size = '3'>**The Law of Large Numbers**<br /><br />
The law of large numbers (LLN) states that as sample size grows, the sample mean gets closer to the population mean<br /><br />
A die can take on the values 1, 2, 3, 4, 5, 6; if a die is "fair" each value has a 1-in-6 chance of coming   
The expected value of a roll of a die is (1 + 2 + 3 + 4 + 5 + 6)/6 = 3.5  
The LLN states that with more trials (rolls of the die), the mean of the trials will converge to 3.5

```python
# Import required resources

import random                      # Produces random numbers
import matplotlib.pyplot as plt    # Plot graphs
import numpy as np                 # Creates arrays and matrices and manipulation them
```

```python
# Set the seed for the random number generator - fixing the seed allow replication

random.seed(5623)
```

<font color = 'blue' size = '3'>**Define a Function to Simulate Rolling a Die**

```python
def Roll(n):
    """This function simulates rolling a die n-times"""

    result = []  # Creates an empty array (vector) to store the results of rolling the die
    
    for i in range(1, n + 1):
        result.append(random.choice([1, 2, 3, 4, 5, 6]))
    return result
```

<font color = 'blue' size = '3'>Invoke Roll(n) for n = 100 and visualize how the mean of the rolls changes as the number of rolls increases<br /><br />
Remember, average of rolled values should converge to the expected value (3.5)

```python
result = Roll(100)
```

```python
averages1 = []   # Create an list to store the values of averages

_cumsum = np.cumsum(result)  # Cumulative sum of rolled values

for index in range(len(_cumsum)):
    averages1.append(_cumsum[index]/(index+1))
```

```python
# Plot the averages

plt.figure(figsize=(14, 6))  # Set figure size (width, height)

plt.plot(averages1)

plt.title('\nAverage Value of Rolls of a Die = 100\n', fontsize = 18)  # Plot title
plt.xlabel('Trials', fontsize = 16, loc = 'right')  # Label the x-axis
plt.ylabel('Average of all outcomes', fontsize = 16, loc = 'top')   # Label the y-axis 
 
plt.show()
```
