## <font color = blue> Demand and Supply

```python
# Load resources

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
# Define the demand and supply curves

def D(q):
    return (q - 20)**2

def S(q):
    return (q**2)

q = np.linspace(0, 16, 1000)  # Generate data on quantity
```

```python
# Plot the demand and supply curves

plt.figure(figsize= (8, 6))

plt.plot(q, D(q), label = 'Demand Curve', color = 'darkorange')  # Plot demand curve

plt.plot(q, S(q), label = 'Supply Curve', color = 'darkgreen')  # Plot supply curve

plt.title('Demand and Supply\n', color = 'blue', fontsize = 20)
plt.legend(frameon = False, fontsize = 14)
plt.xlabel('\nQuantity ($q$)', fontsize = 18)
plt.ylabel('Price ($p$)\n', fontsize = 18)

plt.show()
```

## <font color = blue> Equilibrium Price  
    
    Equilibrium price is determined by the intersection of demand and supply   

    To find equilibrium price solve for $S$($q$) = $D$($q$)  
    
    This can be solved with Sympy

```python
# Solve for equilibrium price using Sympy

q = sy.Symbol('q')  # Set the variable to solve for

eq = sy.Eq(S(q), D(q))  # System of two equations

p = sy.solve(eq)  # Solve for equil. price

print('\nEqualibrium price =', p, '\n')
```

```python
# Assure that this is the equil price by solving for QD and QS

S(10)
```

```python
# Add the equilibrium to the demand and supply plot

plt.figure(figsize= (8, 6))

q = np.linspace(0, 16, 1000)

plt.plot(q, D(q), label = 'Demand Curve', color = 'darkorange')  # Plot demand curve

plt.plot(q, S(q), label = 'Supply Curve', color = 'darkgreen')  # Plot supply curve

plt.plot(10, 100, 'o', markersize = 12, color = 'red')  # Plot the equilibrium points  # Indicate equilibrium with a marker

plt.margins(0)  # Force axes to start from (0,0)

plt.hlines(y = 100, xmin = 0, xmax = 10, ls = ':', lw = 2, color = 'red')  # Add horizontal line from y axis to equil pt
plt.vlines(x = 10, ymin = 0, ymax = 100, ls = ':', lw = 2, color = 'red')  # Add vertical line from x axis to equil pt

plt.title('Demand and Supply', color = 'blue', fontsize = 20)
plt.legend(frameon = False, fontsize = 14)
plt.xlabel('\nQuantity ($q$)', fontsize = 18)
plt.ylabel('Price ($p$)\n', fontsize = 18)

plt.annotate('$p*$ = 100, $q*$ = 10',  xy = (10, 100), xytext = (10.5, 98), fontsize = 16)  # Label equil point

plt.show()
```
